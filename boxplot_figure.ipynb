```python
import pandas as pd
import matplotlib.pyplot as plt
font_size = 11
plt.rcParams.update({'font.size': font_size})
```

```python
#wgan
df=pd.read_csv('./logs/wgan_result_match_top3.csv', sep=',', header=None)
df = df.T
df = pd.DataFrame(df.values[1:], columns= df.iloc[0])
df /= 10
boxplot = df.boxplot()
boxplot.set_xlabel('wavelength (nm)')
boxplot.set_ylabel('percentage')
boxplot.set_ylim([0, 1])
plt.savefig("./logs/result/wgan_result_match_top3.png", format="png", dpi=300)
plt.show()
plt.clf()
# wgan truth
df=pd.read_csv('./logs/wgan_result_truth.csv', sep=',', header=None)
# df = df.T
df = pd.DataFrame(df.values[:], columns= df.iloc[0])
df /= 120
boxplot = df.boxplot()
boxplot.set_xlabel('wavelength (nm)')
boxplot.set_ylabel('accuracy')
boxplot.set_ylim([0, 0.4])
plt.savefig("./logs/result/wgan_result_truth.png", format="png", dpi=300)
plt.show()
```
